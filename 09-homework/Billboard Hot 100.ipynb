```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
```

```python
ranks = doc.find_all(class_='chart-list-item__rank')
for rank in ranks:
    print(rank.text)
```

```python
artists = doc.find_all(class_='chart-list-item__artist')
for artist in artists:
    print(artist.text)
```

```python
songs = doc.find_all(class_='chart-list-item__title-text')
for song in songs:
    print(song.text)
```

```python
charts = doc.find_all(class_='chart-list-item__first-row chart-list-item__cursor-pointer')

for chart in charts:
    row = {}
    
    row['rank'] = chart.find(class_='chart-list-item__rank').text.strip()
    row['artist'] = chart.find(class_='chart-list-item__artist').text.strip()
    row['song'] = chart.find(class_='chart-list-item__title-text').text.strip()

    print(row)
```

```python
rows = []

charts = doc.find_all(class_='chart-list-item__first-row chart-list-item__cursor-pointer')

for chart in charts:
    row = {}
    
    row['rank'] = chart.find(class_='chart-list-item__rank').text.strip()
    row['artist'] = chart.find(class_='chart-list-item__artist').text.strip()
    row['song'] = chart.find(class_='chart-list-item__title-text').text.strip()

    rows.append(row)
    
rows

df = pd.DataFrame(rows)
df
```
