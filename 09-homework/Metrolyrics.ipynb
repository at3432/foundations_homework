```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://www.metrolyrics.com/rem-lyrics.html')
doc = BeautifulSoup(response.text)
```

```python
for music in doc.find_all('tr'):
    music_dict = {}
    
    if len(music.find_all('td')) > 0:
        music_dict['title'] = music.find_all('td')[1].text
        music_dict['URL'] = music.find('a')['href']
        music_dict['year'] = music.find_all('td')[2].text
        music_dict['popularity'] = music.find('span')
    else:
        pass
    
    print(music_dict)

    
```

```python
rows = []
for music in doc.find_all('tr'):
    music_dict = {}
    
    if len(music.find_all('td')) > 0:
        music_dict['title'] = music.find_all('td')[1].text.strip('\n\n').replace('Lyrics','')
        music_dict['URL'] = music.find('a')['href']
        music_dict['year'] = music.find_all('td')[2].text
        music_dict['popularity'] = music.find('span').select('span[style]')[0]['style'][6:-2]
    else:
        pass

    rows.append(music_dict)
print(rows)   

```

```python
df = pd.DataFrame(rows)
```

```python
df = df[['title', 'year','popularity', 'URL']]
```

```python
df.to_csv('REM1.csv', index=False)
```

```python
links = df.dropna()['URL']
```

```python
links[1]
```

```python
response = requests.get(links[1])
doc = BeautifulSoup(response.text)
```

```python
lyric1 = doc.find_all('p', attrs={'class': 'verse'})
```

```python
for lyric in lyric1:
    print(lyric.text)
```

```python
all_lyrics = []
for link in links:
    response = requests.get(link)
    doc = BeautifulSoup(response.text)
    lyric = doc.find_all('p', attrs={'class': 'verse'})
    lyric_content = str()
    for lyric in lyric:
        lyric_content=lyric_content+lyric.text
    all_lyrics.append(lyric_content)
    
```

```python
len(all_lyrics)
```

```python
df = df.dropna()
```

```python
df['lyrics']=all_lyrics
```
