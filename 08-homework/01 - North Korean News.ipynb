```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf")
doc = BeautifulSoup(response.text, 'html.parser')

# http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf;jsessionid=FBF0A38A6F09F58025D4F00BA1BCABF7 ---> 
# this was working in English but the next day when I checked it again it returned the results in Korean
# however they are correct
```

```python
NKrows = []
for headline in headlines:
    NKrow = {}
    
    NKrow['headline'] = headline.text
    NKrow['onclick'] = headline.attrs['onclick']
    NKrow['ID'] = headline.attrs['onclick'][16:25]

    NKrows.append(NKrow)
NKrows    

```

```python
import pandas as pd
df = pd.DataFrame(NKrows)
df.head()
```

```python
df.to_csv('nk-news.csv', index=False)
```
