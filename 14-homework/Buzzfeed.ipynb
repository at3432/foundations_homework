```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv("data/gender-by-quarter.csv")
df.head()
```

```python
df2 = pd.read_csv("data/diversity-by-quarter.csv")
df2.head()
```

```python
df.pct_female = df.pct_female.astype(int)
```

```python
df3 = df[df.toa_group == 'all-permanent']
df3.head()
```

```python
df4 = df[df.toa_group == 'midlevel-appointees']
df4.head()
```

```python
merged_df = df3.merge(df4, on='quarter')
merged_df.head()
```

```python
merged_df.plot(x='quarter', y=['pct_female_y', 'pct_female_x'], ylim=[0, 60], figsize=(8, 6), color=['red', 'black'],
         marker='o', mfc='white', grid=True, title='% of Federal Employees Identifying as Female', legend=False)

plt.savefig("Buzzfeed1.pdf")
```
